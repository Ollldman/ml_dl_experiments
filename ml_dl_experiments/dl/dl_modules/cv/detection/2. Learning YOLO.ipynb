```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x: np.ndarray = np.linspace(-10, 10, 100)
```

```python
def mish(x: np.ndarray) -> np.ndarray:
    x = np.clip(x, -20, 20)
    return x * np.tanh(np.log(1 + np.exp(x)))
```

```python
y: np.ndarray = mish(x)
```

```python
from ml_dl_experiments.dl import func_activation


y_two: np.ndarray = func_activation.swish(x)
```

```python
fig, ax = plt.subplots(figsize=(5, 5))
ax.plot(x, y, label="Mish", c='g')
ax.plot(x, y_two, label="Swish", c='r')
plt.suptitle("Mish vs Swish")
plt.ylim((-0.5, 4))
plt.xlim((-7, 7))
plt.legend()
plt.style.use("ggplot")
plt.show()
```
